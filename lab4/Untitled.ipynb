```python
import numpy as np
from pyDatalog import pyDatalog
from pyDatalog import pyEngine

def maxOfProduct(x, y):
    z = []
    for x1 in x:
        for y1 in y.T:
            z.append(max(np.multiply(x1, y1)))
    return np.array(z).reshape((x.shape[0], y.shape[1]))

r1 = np.array([[0, 0.2, 0.7, 0.8], [0, 0.3, 0.7, 0.5], [0.5, 0.5, 0.5, 0.5], [0.7, 0.4, 0.4, 0.6], [0.7, 0.7, 0.3, 0.3]])
r2 = np.array([[0, 0.3, 0.3, 0.7, 0], [0.3, 0.1, 0.7, 0.3, 0.8], [0.8, 0.6, 0.5, 0.4, 0.3], [0.8, 0.8, 0.5, 0.6, 0.2]])

print("Армия колдунов: \n")
print("Скелеты: ", r1[0])
print("Зомби: ", r1[1])
print("Вампиры: ", r1[2])
print("Ведьмы: ", r1[3])
print("Костяной дракон: ", r1[4], "\n")

print("Армия волшебников: \n")
print("Эф. Магия: ", r2[0])
print("Эф. Физ. Атак: ", r2[1])
print("Вос. Магия: ", r2[2])
print("Вос. Физ. Атак: ", r2[3], "\n")

print("Эффективность способностей(столбцы - волшебники, строки - колдуны): \n")
print(maxOfProduct(r1, r2))
```
